```python
import base64
import sys
from google.oauth2 import service_account
from google.auth.transport.requests import AuthorizedSession

KEY_PATH = "/content/my-key-tts-gcloud.json"
OUT_FILE = "gCloud-Male.wav"
TEXT = """கணினி என்பது கணக்கீடுகள் தரவு சேமிப்பு மற்றும் தகவல் செயலாக்கம் செய்யும் மின்சாதனம் ஆகும் இது மனிதர்களின் உழைப்பை குறைத்து தகவலை துல்லியமாகவும் வேகமாகவும் கையாளுகிறது
கணினி தரவைப் பெற்றுக் கொண்டு அதை செயல்படுத்தி வெளியீட்டாக முடிவை வழங்கும் மின்சாதனம்
கணினி மூன்று கட்டங்களில் செயல்படுகிறது உள்ளீடு செயலாக்கம் வெளியீடு இதன் மூலம் கணக்கீடுகள் தரவு நிர்வாகம் மற்றும் தகவல் பரிமாற்றம் சுலபமாகின்றன

கணினி பல கூறுகளைக் கொண்டுள்ளது அதில் உள்ளீடு சாதனங்கள் நினைவகம் செயலி மற்றும் வெளியீடு சாதனங்கள் முக்கிய பங்காற்றுகின்றன
உள்ளீடு செயலி நினைவகம் மற்றும் வெளியீடு ஆகியவை கணினியின் முக்கிய கூறுகளாகும்
இந்த கூறுகள் ஒருங்கிணைந்து கணினி செயல்பாட்டை மேற்கொள்கின்றன உள்ளீடு சாதனங்கள் தரவைப் பெற செயலி செயல்படுத்தி நினைவகம் சேமித்து வெளியீடு முடிவை காட்டுகிறது

"""

creds = service_account.Credentials.from_service_account_file(
    KEY_PATH,
    scopes=["https://www.googleapis.com/auth/cloud-platform"]
)
authed_session = AuthorizedSession(creds)

url = "https://texttospeech.googleapis.com/v1beta1/text:synthesize"
payload = {
    "input": {"text": TEXT},
    "voice": {
        "languageCode": "ta-IN",
        "modelName": "gemini-2.5-pro-tts",
        "name": "Achernar"
    },
    "audioConfig": {
        "audioEncoding": "MP3",
        "speakingRate": 1.0,
        "pitch": 0
    }
}

resp = authed_session.post(url, json=payload)
if resp.status_code != 200:
    print("Error:", resp.status_code)
    print(resp.text)
    sys.exit(1)

data = resp.json()
audio_b64 = data.get("audioContent")
if not audio_b64:
    print("No audioContent in response:", data)
    sys.exit(1)

with open(OUT_FILE, "wb") as f:
    f.write(base64.b64decode(audio_b64))

print("Saved:", OUT_FILE)

from IPython.display import Audio, display

# Display an audio player
display(Audio("gCloud-Male.wav", autoplay=False))
```
